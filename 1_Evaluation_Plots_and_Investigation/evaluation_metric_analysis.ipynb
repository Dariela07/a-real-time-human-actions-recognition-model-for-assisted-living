### Adjusting Confusion Matrix format

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Confusion matrix (example matrix based on your description)
confusion_matrix = np.array([
    [117,   0,   0,   1],  # Falling
    [  1, 116,   0,   1],  # Staggering
    [  0,   0, 100,  18],  # Chest Pain
    [  1,  15,  18, 366]   # Others
])

# Class labels
class_labels = ['Falling', 'Staggering', 'Chest Pain', 'Others']

# Create a heatmap
plt.figure(figsize=(10, 10))
ax = sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Blues", 
                 xticklabels=class_labels, yticklabels=class_labels, 
                 cbar=True, annot_kws={"size": 14}, cbar_kws={"shrink": 0.8})

ax.set_aspect('equal')
# Add labels, title, and adjust layout

# Customize color bar font size
cbar = ax.collections[0].colorbar  # Get the color bar
cbar.ax.tick_params(labelsize=12)  # Increase font size of color bar tick labels


ax.set_xlabel('Predicted Labels', fontsize=18)
ax.set_ylabel('True Labels', fontsize=18)
ax.set_title('I3D Model Test Confusion Matrix', fontsize=20) # , fontweight='bold'
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14, rotation=0)

# Show the heatmap
plt.tight_layout()
plt.savefig('confusion_matrix_i3d_test3.png', dpi=300)  # Save as PNG with high resolution
plt.show()
```
